```python
import numpy as np
import csv
import os
import matplotlib.pyplot as plt
import pandas as pd
import math
import scipy
# import scipy.signal
```

```python
'''Open the csv file and plot the first and last frames of the data. Index -1 plots last frame in a data set'''

with open('/Volumes/GoogleDrive/Shared drives/Wellesley Solar/Current Projects/SSRL January 2020/XRay_CSV files/B1_MAPbI2Br_Dark60min.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    x=[]
    y=[]
    z=[]
    for row in csv_reader:
        if line_count == 0:
#             print(f'Column names are {", ".join(row)}')
            line_count += 1
        else:
            x.append(float(row[0]))
            y.append(float(row[1]))
            z.append(float(row[-1]))
            line_count += 1
plt.plot(x,y, marker='o', color='blue')
plt.plot(x,z, marker='o', color='red')
plt.title('Initial:')
plt.xlabel('2-theta')
plt.ylabel('Intensity')
plt.show()
#print(f'Processed {line_count} lines.')
#add rolling ball background subtraction
#import cv2
#from cv2_rolling_ball import subtract_background_rolling_ball
#img = cv2.imread(f'path to img', 0)
#img, background = subtract_background_rolling_ball(img, 23, light_background=True,use_paraboloid=False, do_presmooth=True)
#from scipy.signal import medfilt
#print(medfilt([x,y], 25))
```

```python
'''Converting the 2-theta data into Q'''

import math
print(math.pi)
q =[4*math.pi*math.sin(math.pi/180*row/2)/0.9763 for row in x]
plt.plot(q,y, marker='.',color='blue')
plt.title('Initial:')
plt.xlabel('Q')
plt.ylabel('Intensity')
plt.show()
```

```python
'''Set data limits '''
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return array[idx]

q_1 = 0.92
q_2 = 1.12
limit1 = q.index(find_nearest(q, q_1))
limit2 = q.index(find_nearest(q, q_2))

q_sub = q[limit1:limit2]
int_sub = y[limit1:limit2]
plt.plot(q_sub,int_sub)

```

```python
'''Remove background '''
slope = (np.mean(y[limit2-10:limit2])-np.mean(y[limit1:limit1+10]))/(np.mean(q[limit2-10:limit2])-np.mean(q[limit1:limit1+10]))

b = y[limit1]-slope*q[limit1]
back = [slope*element+b for element in q]

int_correct = np.array(int_sub)-np.array(back[limit1:limit2])

plt.plot(np.array(q_sub),int_correct)
```

```python
'''gaussian fitting '''
from scipy import optimize

def gaussian(t, height, center, width):
    return height*np.exp(-(t - center)**2/(2*width**2)) 

def two_gaussians(t, h1, c1, w1, h2, c2, w2):
    return (gaussian(t, h1, c1, w1) +
        gaussian(t, h2, c2, w2))

errfunc2 = lambda p, t, e: (two_gaussians(t, *p) - e)**2

guess2= [25, 1.01, .05, 25, 1.03, .05]
optim2, success = optimize.leastsq(errfunc2, guess2[:], args=(np.array(q_sub), int_correct))

print(optim2[0],optim2[1],optim2[2]/2)
print(optim2[3],optim2[4],optim2[5]/2)


plt.scatter(np.array(q_sub), int_correct, c='pink', label='measurement', marker='.', edgecolors=None)
plt.plot(np.array(q_sub), two_gaussians(np.array(q_sub), *optim2))
    
```
